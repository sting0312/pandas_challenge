```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#total number of schools- count of unique school IDS
school_names = school_data_complete["school_name"].unique()
school_count = len(school_names)

#Total number of students- count of school ID's/length of column
total_students = len(school_data_complete["Student ID"])

#total budget - sum of all 15 id's budget
school_budget = school_data_complete["budget"].unique()
total_budget = sum(school_budget)

#average math score
avg_math = school_data_complete["math_score"].mean()

#average reading score
avg_read = school_data_complete["reading_score"].mean()

grades_bin = [0,69,100]
grades_labels= [0, 1]
passed_math = pd.cut(school_data_complete["math_score"], grades_bin, labels= grades_labels)
school_data_complete["Passed Math"]= passed_math
passed_read = pd.cut(school_data_complete["reading_score"], grades_bin, labels= grades_labels)
school_data_complete["Passed Reading"]= passed_read
school_data_complete[["Passed Math", "Passed Reading"]] = school_data_complete[["Passed Math", "Passed Reading"]].apply(pd.to_numeric)
pass_math = school_data_complete["Passed Math"].sum()
pass_read = school_data_complete["Passed Reading"].sum()
percent_math = pass_math/total_students
percent_read = pass_read/total_students
passed_all = school_data_complete[(school_data_complete["Passed Math"] == 1) & (school_data_complete["Passed Reading"] ==1)]
percent_all = len(passed_all)/total_students
```

```python
#Summary Table
summary_complete = pd.DataFrame({"Total Number of Schools": [school_count],
                                    "Total Number of Students": [total_students],
                                    "Total Budget of Schools": [total_budget],
                                    "Average Math Score": [avg_math],
                                    "Average Reading Score": [avg_read],
                                    "Percent of Students Passing Math": [percent_math],
                                    "Percent of Students Passing Reading": [percent_read],
                                    "Overall Passing Percent": [percent_all]
                                })
summary_complete
```

## School Summary

```python
school_data_complete["Percent Passing Math"] = school_data_complete["math_score"] >= 70
school_data_complete["Percent Passing Reading"] = school_data_complete["reading_score"] >= 70
school_data_complete["Percent Overall Passing"] = school_data_complete["Percent Passing Math"] & school_data_complete["Percent Passing Reading"]
school_data_complete.head()
```

```python
school_data_complete2 = school_data_complete.rename(columns={
    "Student ID": "Student ID",
    "student_name": "Student Name",
    "gender": "Student Gender",
    "grade": "Grade",
    "school_name": "School Name",
    "reading_score": "Reading Score",
    "math_score": "Math Score",
    "School ID": "School ID",
    "type": "School Type",
    "size": "School Size",
    "budget": "School Budget",
    
})
perstudentbudget = school_data_complete2["School Budget"] / school_data_complete2["School Size"]
school_data_complete2["Per Student Budget"] = perstudentbudget
school_data_complete2
```

```python
grouped_by_name = school_data_complete2.groupby(["School Name"])

#grouped_by_name.mean()
```

```python
studentcount = grouped_by_name["Student ID"].count()

schooltype = grouped_by_name["School Type"].unique()

schoolsize = grouped_by_name["School Size"].unique()

schoolbudget = grouped_by_name["School Budget"].unique()

perstudentbudget = grouped_by_name["Per Student Budget"].unique()

averagemath = grouped_by_name["Math Score"].mean()

averageread = grouped_by_name["Reading Score"].mean()

mathpass = grouped_by_name["Percent Passing Math"].mean() *100

readpass = grouped_by_name["Percent Passing Reading"].mean() *100

overallpass = grouped_by_name["Percent Overall Passing"].mean() *100
```

```python
school_summary = pd.DataFrame({
    "School Type": schooltype,
    "School Size": schoolsize,
    "Total Students": studentcount,
    "Total School Budget": schoolbudget,
    "Per Student Budget": perstudentbudget,
    "Average Math Score": averagemath,
    "Average Reading Score": averageread,
    "% Passing Math": mathpass,
    "% Passing Reading": readpass,
    "% Overall Passing": overallpass
})
school_summary
```

```python
scores_data = school_data_complete2[["School ID", 
                                                           "School Name",
                                                           "School Type",
                                                           "School Size",
                                                           "School Budget",
                                                           "Per Student Budget",
                                                           "Student ID",
                                                           "Student Name", 
                                                           "Student Gender", 
                                                           "Grade", 
                                                           "Reading Score",
                                                           "Math Score",
                                                           "Percent Passing Reading",
                                                           "Percent Passing Math",
                                                           "Percent Overall Passing" ]]
```

```python
refined_scores = scores_data[["Per Student Budget",
                                "School Type",
                                "Student ID",
                                "Reading Score",
                                "Math Score",
                                "% Passing Reading",
                                "% Passing Math",
                                "% Overall Passing" ]]
refined_scores
```

```python
bins = [575,594.9,614.9,634.9,656]
labels = ["<595","595-615","615-635","635-655",]

refined_scores["Per Student Budget"] = pd.cut(scores_data["Per Student Budget"], bins, labels=labels)
refined_scores

scores_refined_budget = refined_scores.groupby("Per Student Budget")
#scores_refined_budget.mean()

mathaverage = scores_refined_budget["Math Score"].mean()
readaverage = scores_refined_budget["Reading Score"].mean()
mathpass = scores_refined_budget["% Passing Math"].mean()*100
readpass = scores_refined_budget["% Passing Reading"].mean() *100
mathreadpass = scores_refined_budget["% Overall Passing"].mean() *100

scores_budget_summary = pd.DataFrame({"Average Math Score" : mathaverage,
                                   "Average Reading Score": readaverage,
                                   "% Passing Math": mathpass,
                                   "% Passing Reading": readpass,
                                   "% Overall Passing" : mathreadpass })
scores_budget_summary
```
